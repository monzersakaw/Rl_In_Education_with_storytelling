```python
import random

class MLModel:
    def __init__(self):
        self.operations = ['+', '-', '*', '/']

    def generate_question(self, level):
        if level == 'easy':
            a, b = random.randint(1, 10), random.randint(1, 10)
        elif level == 'medium':
            a, b = random.randint(10, 99), random.randint(10, 99)
        elif level == 'hard':
            a, b = random.randint(100, 999), random.randint(100, 999)


        op = random.choice(self.operations)
        if op == '+':
            question = f"{a} + {b}"
            answer = a + b
        elif op == '-':
            question = f"{a} - {b}"
            answer = a - b
        elif op == '*':
            question = f"{a} * {b}"
            answer = a * b
        elif op == '/':
            if b == 0:
                b = 1
            question = f"{a * b} / {b}"
            answer = a
        return question, answer
    def provide_explanation(self, question, correct_answer):
        if '+' in question:
            a, b = map(int, question.split(' + '))
            explanation = (
                f"To solve {a} + {b}:\n"
                f"1. Start with the number {a}.\n"
                f"2. Add {b} to {a}.\n"
                f"3. The result is {a} + {b} = {correct_answer}.\n"
                f"For example, if you have {a} apples and you get {b} more apples, you will have {correct_answer} apples."
            )
        elif '-' in question:
            a, b = map(int, question.split(' - '))
            explanation = (
                f"To solve {a} - {b}:\n"
                f"1. Start with the number {a}.\n"
                f"2. Subtract {b} from {a}.\n"
                f"3. The result is {a} - {b} = {correct_answer}.\n"
                f"For example, if you have {a} candies and you give away {b} candies, you will have {correct_answer} candies left."
            )
        elif '*' in question:
            a, b = map(int, question.split(' * '))
            explanation = (
                f"To solve {a} * {b}:\n"
                f"1. Take {a} and multiply it by {b}.\n"
                f"2. The result is {a} * {b} = {correct_answer}.\n"
                f"For example, if you have {a} bags of apples and each bag contains {b} apples, you will have {correct_answer} apples in total."
            )
        elif '/' in question:
            a, b = question.split(' / ')
            a, b = int(a), int(b)
            explanation = (
                f"To solve {a} / {b}:\n"
                f"1. Take {a} and divide it by {b}.\n"
                f"2. The result is {a} / {b} = {correct_answer}.\n"
                f"For example, if you have {a} Egyptian LE and you want to divide them equally among {b} people, each person will get {correct_answer} Egyptian LE."
            )
        return explanation
```

```python
class StudentModel:
    def __init__(self):
        self.correct_answers = 0
        self.total_questions = 0

    def answer_question(self, question):
        try:
            user_input = input(f"Solve the following question: {question} = ")
            return float(user_input)
        except ValueError:
            print("Invalid input. Please enter a number.")
            return None

    def provide_feedback(self, is_correct):
        self.total_questions += 1
        if is_correct:
            self.correct_answers += 1
            print("Correct!")
        else:
            print("Incorrect.")

    def get_score(self):
        return self.correct_answers, self.total_questions
```

```python
import random
import numpy as np

# Generate a mix of easy and medium-level questions
easy_questions = [f"{random.randint(1, 10)} + {random.randint(1, 10)}" for _ in range(5)]
medium_questions = [f"{random.randint(10, 99)} - {random.randint(10, 99)}" for _ in range(5)]

# Combine questions and assign binary responses (0 for incorrect, 1 for correct)
# For demonstration purposes, let's assume the student answered all questions correctly
questions = easy_questions + medium_questions
responses = [1] * 5 + [1] * 5  # Assuming all correct answers

# Shuffle the data
data = list(zip(questions, responses))
random.shuffle(data)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Extract questions and responses from data
questions, responses = zip(*data)

# Tokenize questions
tokenizer = Tokenizer()
tokenizer.fit_on_texts(questions)
question_sequences = tokenizer.texts_to_sequences(questions)
max_len = max(len(seq) for seq in question_sequences)

# Pad sequences to ensure uniform input size
padded_sequences = pad_sequences(question_sequences, maxlen=max_len)

# Define the neural network model
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(len(tokenizer.word_index) + 1, 64),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(padded_sequences, np.array(responses), epochs=10, batch_size=2)
```

```python
# Example student responses
student_responses = ["5 + 3", "10 - 2", "20 + 30", "50 - 25", "90 - 40", "12 + 8", "85 - 20", "70 + 15", "15 * 5", "50 / 5"]

# Convert student responses to tokenized sequences
student_sequences = tokenizer.texts_to_sequences(student_responses)
padded_student_sequences = pad_sequences(student_sequences, maxlen=max_len)

# Predict proficiency level
predicted_probabilities = model.predict(padded_student_sequences)

# Assuming threshold for determining proficiency level is 0.5
proficiency_levels = ['Easy' if prob < 0.5 else 'Medium' for prob in predicted_probabilities]

# Determine overall proficiency level based on majority vote
overall_proficiency = max(set(proficiency_levels), key=proficiency_levels.count)
```

```python
if overall_proficiency == 'Easy':
    level = 'easy'
elif overall_proficiency == 'Medium':
    level = 'medium'
```

```python
import random

def main():
    ml_model = MLModel()
    student = StudentModel()
    print("welcome to Monsters Inc for math reinforcement in this Application you are asked math questions and if you answered correctly you will unlock story parts other wise the monsters will be anggry at you:")

    # Start with the easy level
    level = 'medium'
    prev_level = 'easy'

    # Story parts
    positive_story = [
        "In the heart of Monstropolis, where laughter fuels the city, lies Monsters, Inc., a vibrant factory bustling with energy and excitement. At its helm are Sulley and Mike, an unlikely duo whose laughter-inducing antics keep the factory running smoothly. With Sulley's towering presence and Mike's infectious enthusiasm, they bring joy to all who cross their path, from fellow monsters to the curious human child, Boo.You're doing great! Boo is cheering for you!",
        "As Sulley and Mike embark on their adventure with Boo, they discover a newfound sense of purpose and companionship. Despite the challenges they face, their journey is filled with heartwarming moments and unexpected friendships. Whether they're dodging the schemes of their rival Randall or receiving unexpected help from Roz, they tackle each obstacle with unwavering optimism and determination.The monsters are amazed by your quick thinking!",
        "Through their laughter-filled escapades, Sulley, Mike, and Boo teach us that friendship knows no bounds and that even the scariest of situations can be overcome with a smile. As they bring laughter and joy to the world around them, they remind us that sometimes the greatest adventures are the ones shared with those we love. In the end, Monsters, Inc. isn't just a factory; it's a place where laughter reigns supreme and where every day is filled with endless possibilities and boundless, cheerYou've impressed the top scarers with your skills!."

    ]

    negative_story = [
        "Oh no! The monsters are getting anxious...",
        "The monsters are worried about your progress...",
        "It seems like the monsters are losing hope..."
    ]

    while True:
        print(f"\nLevel: {level}")
        num_questions = 5  # Number of questions to ask
        questions_and_answers = []  # Store questions, correct answers, and student answers

        for _ in range(num_questions):
            question, correct_answer = ml_model.generate_question(level)
            student_answer = student.answer_question(question)

            if student_answer is not None:
                is_correct = (student_answer == correct_answer)
                student.provide_feedback(is_correct)
                questions_and_answers.append((question, correct_answer, student_answer, is_correct))

        correct, total = student.get_score()
        print(f"\nFinal Score: {correct} out of {total}")

        # Display positive or negative story based on performance
        if correct >= 3:
            if level == 'easy':
                print("Congratulations! You've passed the easy level. Moving to the medium level.")
                prev_level = 'easy'
                level = 'medium'
                print(random.choice(positive_story))
                student.correct_answers = 0  # Resetting the score for the new level
                student.total_questions = 0
            elif level == 'medium':
                print("Congratulations! You've passed the medium level. Moving to the hard level.")
                prev_level = 'medium'
                level = 'hard'
                print(random.choice(positive_story))
                student.correct_answers = 0  # Resetting the score for the new level
                student.total_questions = 0
            elif level == 'hard':
                print("Congratulations! You've passed the hard level.")
                print("You've saved Monster Inc. from disaster!")
                break
        else:
            print("\nIt seems you need some help with the questions. Here are the explanations for the incorrect answers:")
            for question, correct_answer, student_answer, is_correct in questions_and_answers:
                if not is_correct:
                    explanation = ml_model.provide_explanation(question, correct_answer)
                    print(f"\nQuestion: {question}")
                    print(f"Your answer: {student_answer}")
                    print(f"Correct answer: {correct_answer}")
                    print(f"Explanation: {explanation}")
            print(random.choice(negative_story))

            # Return to the previous level if applicable
            if prev_level:
                print(f"You didn't pass the {level} level. Returning to the {prev_level} level.")
                if level == 'hard':
                    level = 'medium'
                elif level == 'medium':
                    level = 'easy'
                prev_level = 'easy'
                student.correct_answers = 0  # Resetting the score for the new level
                student.total_questions = 0
                continue  # Restart the loop to ask questions in the previous level

if __name__ == "__main__":
    main()
```
